# selecting data with SQLite (aka don't use pandas) #

- note, I don't actually use pandas
- pandas is very popular, so I figured I would give it due credit
  - but, I prefer to use relational syntax and semantics (i.e., databases)
  
- soapbox (my opinion)
  - data frame syntax (also in R, Python Blaze) is horrible and no one should use it
    - it uses index semantics, like indexing arrays
  - learn SQL instead, it
    - teaches you the relational data model
    - the syntax isn't ugly and unwieldly (strong opinion)
    - teaches you the same concepts as map-reduce (Hadoop)
    - teaches you data parallelism
    - implementation of joins are sane and optimized 
    - it is streaming, out-of-core
    - it is functional
    - it is nearly universal
    - SQLite provides virtual tables to treat any data source as a relational store without ingest
  - (also don't use key-value stores, but that's a different soapbox)
  
- what I will be showing below is raw SQL syntax strings
  - there are libraries that make it easier to express queries
    in a domain specific language (LINQ for example)
  - I will show one in R later called "dplyr"

```python
for row in cursor.execute('select * from can where Points_0 > 6 and Points_2 > -5'):
    print(row)
```

```python
def query_to_columns(query, cursor):
    cursor.execute(query)
    columns = [[x] for x in cursor.fetchone()]
    for row in cursor.fetchall():
        for col, column in zip(row, columns):
            column.append(col)
    return columns

p0, p1, p2 = query_to_columns('select Points_0, Points_1, Points_2 from can where Points_2 > -6 and Points_2 < -4', cursor)

plot.plot(p0, p2, '.')
```

```python
figure = plot.figure()
axes = figure.add_subplot('111', projection='3d')
axes.scatter(p0, p1, p2)
plot.show()
```

```python
q = 'select count(*) from can where VEL_0 * VEL_0 + VEL_1 * VEL_1 * VEL_2 * VEL_2 > 1e12'
cursor.execute(q).fetchone()
```

```python
q = 'select DISPL_0 * DISPL_0 + DISPL_1 * DISPL_1 + DISPL_2 * DISPL_2 as disp, \
VEL_0 * VEL_0 + VEL_1 * VEL_1 * VEL_2 * VEL_2 as vel, Points_0, Points_1, Points_2 from \
can where vel > 1e12'
disp, vel, x, y, z = query_to_columns(q, cursor)
plot.plot(vel, disp, '.')
```

# adding color, legends and labels #

```python
from matplotlib.colors import Normalize

fig = plot.figure()
axes = fig.add_subplot('111')
im = axes.scatter(x, disp, c=vel, marker='.', cmap=plot.get_cmap('gist_gray'))
axes.set_title('x vs. displacement colors by velocity')
axes.set_xlabel('x')
axes.set_ylabel('displacement')

cax = fig.add_axes([0.15, 0.75, 0.5, 0.05])
cb = plot.colorbar(im, cax, cmap=plot.get_cmap('gist_gray'), 
                   norm=Normalize(vmin=min(vel), vmax=max(vel)), orientation='horizontal')
cax.set_title('velocity')

plot.show()
```

```python
fig = plot.figure()
axes = fig.add_subplot('111')
im = axes.scatter(vel, disp, c=x, marker='.', cmap=plot.get_cmap('inferno'))
axes.set_title('displacement vs. velocity colors by x')
axes.set_xlabel('displacement')
axes.set_ylabel('velocity')

cax = fig.add_axes([0.8, 0.15, 0.05, 0.65])
cb = plot.colorbar(im, cax, cmap=plot.get_cmap('inferno'), 
                   norm=Normalize(vmin=min(x), vmax=max(x)))
cax.set_title('x')

plot.show()
```

# More than just scatterplots #

```python
q = 'select Points_0, abs(DISPL_0)+abs(DISPL_1)+abs(DISPL_2) from can where Points_2 > -4.25 and Points_2 < -3.75'
a, b = query_to_columns(q, cursor)

window = numpy.ones(7)

xs = sorted(a)
ys = [j for (i, j) in sorted(zip(a, b))]

ys = numpy.convolve(window / window.sum(), ys, mode='same')

plot.plot(xs, ys)
plot.xlabel('x between -4.25 < z < -3.75')
plot.ylabel('displacement')
```

# Interactive Visualization with Bokeh #
## Using the Javascript in your browser to interactively plot data from Python ##

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()

f = figure(title='can')
f.circle(d['Points:0'], d['Points:2'])
show(f)
```

```python
def rgb(xs, ys, zs):
    minx, maxx = min(xs), max(xs)
    miny, maxy = min(ys), max(ys)
    minz, maxz = min(zs), max(zs)
    normx = 255.0 / (maxx - minx)
    normy = 255.0 / (maxy - miny)
    normz = 255.0 / (maxz - minz)
    def rgb_to_str(r, g, b):
        return "#%02x%02x%02x" % (int(r), int(g), int(b))
    return [rgb_to_str((x - minx) * normx, (y - miny) * normy, (z - minz) * normz) 
            for x, y, z in zip(xs, ys, zs)]
        

f = figure(title='can')
colors = rgb(x, y, z)
f.circle(disp, vel, fill_color=colors, line_color=colors)
show(f)
```

# Scientific Visualization Specific Tools #

- Typical plotting tools are good for the majority of visualization needs
- But what if
  - The data is 3D
  - The data is comprised of a "mesh" (i.e., cells and vertices)
- For 3D and mesh data, game engines are good at this
  - There aren't many open source ones
  - Custom visualization with OpenGL
- Visualization ToolKit (VTK)
  - Open source library started by Kitware in the 1990s
    - Many open source projects leverage VTK
  - Supports numerous visualization and analysis filters
    - Millions of lines of code
  - Programming model is a directed acyclic graph (data flow)
  - Related library, Imaging ToolKit (ITK) has a partner relationship with VTK, but ITK is geared more towards medical imaging

# Large-scale Scientific Visualization #

- For large scale-meshes, there wasn't a good solution until early 2000s
  - ParaView and VisIt
  - They were developed by the Department of Energy (DOE) to fill this gap
    - Pre-cloud computing, so they are primarily based to be run on supercomputers or visualization clusters
    - There is movement towards leveraging web technologies, ParaViewWeb and Cinema for example
  - Client-Server where the server runs on the supercomputer or visualization cluster
    - It is a parallel abstraction over VTK with client-server communication
  - They are both built on the open source Visualization ToolKit (VTK)
- There are other tools, such as Mayavi but ParaView and VisIt have the largest footprint in the DOE

# Note on OpenGL and Scientific Visualization #

- Scientific visualization typically displays 3D data (physical phenomena)
  - How does one program 3D graphics?
- OpenGL is the open graphics library that has been the 3D graphics standard since 1991
- It was created by Silicon Graphics (SGI)
  - Fast graphics hardware was coming on the scene
  - There wasn't a good language to abstract the hardware
- It is now the defacto library/language for developing 3D graphics
  - aside from DirectX
- EGL and WebGL dialects are in common use
  - In particular, EGL removes much of the "state machine" complexity of OpenGL
    - Simpler for low-end hardware, but also reflects the current state of graphics hardware
  - WebGL is the dialect that modern web browsers support for 3D rendering
- ParaView leverages OpenGL for the rendering

# Research in Large-scale Scientific Visualization #

- Much of the research into large-scale scientific visualization
  has to deal with the I/O bottleneck at the largest supercomputing scales
  - Exascale Computing Project https://exascaleproject.org
  - I/O is not scaling as fast as compute power
  - This means we can't save data (from a simulation), as often for visualization later
- Primary research thrust in visualization and analysis is 
  - Compute the visualizations & analyses as the simulation is running
    - In situ analysis - tightly coupled
      - Example: ParaView Catalyst http://www.paraview.org/in-situ/, 
    - In transit analysis - loosely coupled
      - Example: ADIOS https://www.olcf.ornl.gov/center-projects/adios/
- Secondary research topics
  - Scalable algorithms
    - Example: VTK-m http://m.vtk.org/index.php/Main_Page
  - Explorable visualization with in situ data products
    - Example: Cinema http://cinemascience.org/
  - Data reduction and compression algorithms (including meta-modeling & statistical sampling)
    - Example: zfp & fpzip http://computation.llnl.gov/projects/floating-point-compression
